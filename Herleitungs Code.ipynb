```python
import pandas as pd
import os

pfad = r"C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt"
dateien = [
    "lineups_deutschland.csv",
    "lineups_england.csv",
    "lineups_frankreich.csv",
    "lineups_italien.csv",
    "lineups_spanien.csv"
]

dataframes = []

for datei in dateien:
    dateipfad = os.path.join(pfad, datei)
    df = pd.read_csv(dateipfad)
    dataframes.append(df)

zusammengefuegt = pd.concat(dataframes, ignore_index=True)

if 'minutes_played' in zusammengefuegt.columns:
    zusammengefuegt['minutes_played'] = (
        zusammengefuegt['minutes_played']
        .fillna(0)  
        .astype(str) 
        .str.replace('.0', '', regex=False)  
        .astype(int) 
    )

output_datei = os.path.join(pfad, "lineups_gesamt.csv")
zusammengefuegt.to_csv(output_datei, index=False)

print(f"Die Dateien wurden erfolgreich zusammengeführt und gespeichert unter: {output_datei}")
```

```python
###häufigste Position berechnen

data_path = r'C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\lineups_gesamt.csv'
df = pd.read_csv(data_path)

position_counts = df.groupby(['player', 'position']).size().reset_index(name='counts')
most_frequent_positions = position_counts.loc[position_counts.groupby('player')['counts'].idxmax()]
most_frequent_positions = most_frequent_positions[['player', 'position']]
most_frequent_positions.columns = ['player', 'most_frequent_position']
most_frequent_positions['most_frequent_position'] = most_frequent_positions['most_frequent_position'].str[:2]

position_mapping = {
    'CB': 'CB',
    'WB': 'WB',
    'LB': 'WB',
    'RB': 'WB',
    'MF': 'MF',
    'DM': 'MF',
    'CM': 'MF',
    'RM': 'WM',
    'LW': 'WM',
    'LM': 'WM',
    'RW': 'WM',
    'AM': 'FW',
    'FW': 'FW'
}

most_frequent_positions['most_frequent_position'] = most_frequent_positions['most_frequent_position'].map(position_mapping)

output_path = r'C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\lineups_gesamt_processed.csv'
most_frequent_positions.to_csv(output_path, index=False)
```

```python
import pandas as pd
import re

final_stats_with_player_df = pd.read_csv(r"C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\final_stats_with_player90.csv")
lineups_gesamt_processed = pd.read_csv(r"C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\lineups_gesamt_processed.csv")
spielerliste_mit_marktwerten = pd.read_csv(r"C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\spielerliste_mit_marktwerten.csv")

lineups_gesamt_processed = lineups_gesamt_processed[lineups_gesamt_processed['most_frequent_position'] != 'GK']

final_stats_with_pos = final_stats_with_player_df.merge(lineups_gesamt_processed[['player', 'most_frequent_position']], on='player', how='inner')

final_stats_with_pos = final_stats_with_pos.drop(columns=['position'], errors='ignore')

final_stats_with_pos = final_stats_with_pos.drop_duplicates(subset='player')

def clean_market_value(value):
    if pd.isna(value):
        return None
    value = re.sub(r'[^0-9,]', '', str(value))
    
    if ',' in value:
        value = value.replace(',', '.')
    else:
        value = '0.' + value

        return float(value)


spielerliste_mit_marktwerten['marktwert'] = spielerliste_mit_marktwerten['marktwert'].apply(clean_market_value)

final_stats_with_pos = final_stats_with_pos.merge(spielerliste_mit_marktwerten[['player', 'marktwert']], on='player', how='left')

final_stats_with_pos.to_csv(r'C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\final_stats_with_position90.csv', index=False)
```

```python
### Marktwerte ermitteln ###

import requests
import pandas as pd
import time

def fetch_market_values(file_path):
    df = pd.read_csv(file_path)
    
    df['marktwert'] = None
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
        "Connection": "keep-alive",
    }
    
    for index, row in df.iterrows():
        player_name = row['player']
        search_url = f"https://www.transfermarkt.de/schnellsuche/ergebnis/schnellsuche?query={player_name}"
        
        print(f"\nVerarbeite Spieler: {player_name}")
        
        time.sleep(2)
        response = requests.get(search_url, headers=headers)

        if ". €" in response.text:
            position = response.text.find(". €")
            start_position = max(position - 10, 0)
            relevant_text = response.text[start_position:position]
            print(f"Text vor '. €': {relevant_text}") 
            
            df.at[index, 'marktwert'] = relevant_text
        else:
            print(f"Begriff '. €' NICHT gefunden für {player_name}")
            df.at[index, 'marktwert'] = "Nicht gefunden"


    output_file = file_path.replace(".csv", "_mit_marktwerten.csv")
    df.to_csv(output_file, index=False)
    print(f"Tabelle mit Marktwerten gespeichert: {output_file}")

file_path = r"C:\Users\paulg\Desktop\DSI - Weiterbildung\Abschlussprojekt\spielerliste.csv"
fetch_market_values(file_path)
```
